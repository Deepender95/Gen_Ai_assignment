```python
#Task 1

# Read input
order_amount_input = input("Enter order amount: ")


if order_amount_input.isdigit():
    order_amount = int(order_amount_input)


    if order_amount >= 2000:
        discount = 0.15
    elif order_amount >= 1500:
        discount = 0.10
    elif order_amount >= 1000:
        discount = 0.07
    else:
        discount = 0.0

    # Calculate final amount
    discount_amount = order_amount * discount
    final_amount = order_amount - discount_amount

    print("Original amount:", order_amount)
    print("Discount applied:", discount * 100, "%")
    print("Final amount:", final_amount)


    tax = final_amount * 0.05
    total = final_amount + tax

    print("Tax 5%:", tax)
    print("Total amount after tax:", total)

else:
    print("Error: Please enter a valid numeric value.")
```

```python
#Task 2

# Given list
orders = [1200, 2500, 800, 1750, 3000]

total_revenue = 0
discounted_orders_count = 0

print("Order summary:")


for order_amount in orders:

    # Apply discount rules
    if order_amount >= 2000:
        discount = 0.15
    elif order_amount >= 1500:
        discount = 0.10
    elif order_amount >= 1000:
        discount = 0.07
    else:
        discount = 0.0

    discount_amount = order_amount * discount
    final_amount = order_amount - discount_amount

    # Count discounted orders
    if discount > 0:
        discounted_orders_count += 1

    # Add to total revenue
    total_revenue += final_amount

    # Print summary row
    print(order_amount, "->", discount * 100, "% ->", final_amount)


print("Total Revenue after discount::: ", total_revenue)
print("Orders with discount::: ", discounted_orders_count)
```

```python
#Task 3

orders = []

while True:
    print("\nMenu:")
    print("1 - Add order amount")
    print("2 - Show orders and totals")
    print("q - Quit")

    choice = input("Enter your choice: ")

    # Quit
    if choice == 'q':
        print("Exiting program...")
        break

    # Add order
    elif choice == '1':
        amount_input = input("Enter order amount: ")

        if amount_input.isdigit():
            orders.append(int(amount_input))
            print("Order added!")
        else:
            print("Invalid input. Please enter a number.")
            continue

    # Show summary
    elif choice == '2':

        if len(orders) == 0:
            print("No orders available.")
            continue

        total_revenue = 0

        print("Order summary:")
        print("-------------------------------")

        for order_amount in orders:

            if order_amount >= 2000:
                discount = 0.15
            elif order_amount >= 1500:
                discount = 0.10
            elif order_amount >= 1000:
                discount = 0.07
            else:
                discount = 0.0

            discount_amount = order_amount * discount
            final_amount = order_amount - discount_amount

            total_revenue += final_amount

            print(order_amount, "->", discount * 100, "% ->", final_amount)

        print("-------------------------------")
        print("Total revenue:::", total_revenue)

    # Invalid choice
    else:
        print("Invalid choice::: ")
        continue
```

```python
#Task 4

daily = [200, 150, 0, 400, 50, -1, 300]

total_sales = 0

for day in daily:

    if day == -1:
        print("Corrupted data found. Stop processing.")
        break

    if day == 0:
        print("No sales for the day. Skipping::: ")
        continue

    total_sales += day
    print("Added:", day, "| Running Total:", total_sales)

print("Final total sales:::", total_sales)
```
