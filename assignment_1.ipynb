```python
#Task 3

#1-----
price_dict = {
    "Apple":120,
    "Milk":40,
    "Tooth brush":30,
    "Book":25,
    "Parle-G": 10,
    "Almonds": 1100,
}
print("price dictionary::: ",price_dict)

#2-----
price_dict["Almonds"] = 900
print("Updated price of almonds::: ",price_dict)
try:
    del price_dict['Book']
    print("deleted book from price dictionary:::",price_dict,)
except KeyError:
    print("Key not found")
    
    
    
#3-----
print("average price of products::: ",sum(price_dict.values())/len(price_dict))

#4-----
print("product with maximum price::: ",max(price_dict,key=price_dict.get))
print("product with minimum price::: ",min(price_dict,key=price_dict.get))


```

```python
#Task 4

#1-----
print("product list",products);
print("categories",categories_list)
print("price dict",price_dict)

catalog = [(x, price_dict.get(x, 0), categories_list[products.index(x)]) for x in products]
print("catelog::: ",catalog)

#2-----
category_to_product =  {}
for x in catalog:
    if x[2] not in category_to_product:
        category_to_product[x[2]] = []
    category_to_product[x[2]].append(x[0])    
print("category to product mapping ::: ",category_to_product)


#3-----
print("category with most products::: ",max(category_to_product, key = lambda x : len(category_to_product[x])))
```
